```python
import pandas as pd
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92],
    'Science': [90, 82, 89],
    'English': [88, 85, 94]
}
 
df = pd.DataFrame(data)
```

```python
df
```

```python
df.melt(id_vars={"Name"}, value_vars= {"Math","Science","English"}, var_name="Subject",value_name="Score")
```

```python
df.pivot(index="Name", columns="Subject", values="Score")
```
